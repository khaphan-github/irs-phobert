# Tiền xử lý dữ liệu tin tức tiếng Việt
## Các bước thực hiện:
1. **Cài đặt thư viện**: Cài đặt các thư viện cần thiết
2. **Tải dữ liệu**: Đọc dữ liệu từ file CSV gốc
3. **Làm sạch dữ liệu**: Loại bỏ dữ liệu thiếu và trùng lặp
4. **Tiền xử lý văn bản**: Chuẩn hóa và tách từ tiếng Việt
5. **Lưu dữ liệu**: Xuất dữ liệu đã xử lý

```python
# Import thư viện pandas để xử lý dữ liệu
import pandas as pd
```

# 1. Tải dữ liệu (Load Data)
Đọc dữ liệu tin tức từ file CSV gốc. Dữ liệu bao gồm các cột: id, title, content, author, topic, source, url, v.v.
Nguồn: https://www.kaggle.com/code/sarahhimeko/vnese-news-classification-textcnn-vdcnn-hybrid?select=Fixed_news_dataset.csv

```python
# Đường dẫn file dữ liệu gốc và file đầu ra
BASE_DATA_DIR = "../data/raw_data/Original_news_dataset.csv"
BASE_OUTPUT_DATA_DIR = "../data/processed_data/Original_news_dataset.csv"
```

```python
# Đọc dữ liệu từ file CSV
df = pd.read_csv(BASE_DATA_DIR, dtype=str, nrows=2000)
df.head()
```

# 2. Làm sạch dữ liệu (Data Cleaning)
Thực hiện các bước làm sạch dữ liệu:
- Loại bỏ các bài báo thiếu tiêu đề hoặc nội dung
- Loại bỏ các bài báo trùng tiêu đề

```python
df = df.dropna(subset=['title', 'content'])
df = df.drop_duplicates(subset=['title'])
df
```

# 3. Tiền xử lý văn bản tiếng Việt (Vietnamese Text Preprocessing)
Sử dụng thư viện `underthesea` để:
- Chuyển văn bản về chữ thường
- Loại bỏ khoảng trắng thừa
- Tách từ tiếng Việt (word tokenization)

Tách từ là bước quan trọng trong xử lý tiếng Việt vì giúp mô hình hiểu được ranh giới từ.

```python
from underthesea import word_tokenize

df["content"] = df["content"].astype(str).str.strip().str.lower()
# df["content"] = df["content"].apply(lambda x: word_tokenize(x, format="text"))
```

```python
df[["id", "content"]].head()
```

# 4. Lưu dữ liệu đã xử lý (Save Processed Data)
Lưu dữ liệu đã được tiền xử lý vào file CSV mới để sử dụng cho các bước tiếp theo.

```python
df.to_csv(BASE_OUTPUT_DATA_DIR, index=False)
```
