```python
import torch
from safetensors.torch import load_file  # Or use torch.load() for .bin files
import numpy as np

# Load weights
weights = load_file("./model/model.safetensors")  # or torch.load("pytorch_model.bin")

# Select the tensor you want (e.g., query layer weight)
tensor = weights["encoder.layer.0.attention.self.query.weight"]  # shape: [768, 768]

# Convert to numpy array
matrix = tensor.cpu().numpy()

# Save as plain text
np.savetxt("query_weight.txt", matrix, fmt="%.6f", delimiter=" ")
```
